### Problem: Largest palindrome product

A palindromic number reads the same both ways. The largest palindrome made from the product of two 2-digit numbers is 9009 = 91 Ã— 99.

Find the largest palindrome made from the product of two 3-digit numbers.

```python
# first define the palindrome function
def palindrome(num1, num2):
    product = num1 * num2
    product_string = str(product)
    rev_product_string = str(product)[::-1] #returns string in reverse
    if product_string == rev_product_string:
        return True
    else:
        return False
```

```python
# now make empty list to house all palindromes
palindrome_list = []
for i in range(100, 1000):
    for j in range(100, 1000):
        if palindrome(i, j) == True:
            palindrome_list.append(i*j)
            #print(str(i) + " times " + str(j) + " equals " + str(i*j))

print("There are " + str(len(palindrome_list)) + " possible palindromes made from the product of two 3-digit numbers.")
print("The largest of these palindrome products is " + str(max(palindrome_list)))
```
